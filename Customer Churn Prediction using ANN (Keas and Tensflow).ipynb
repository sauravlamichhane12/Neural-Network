```python
df.drop(['RowNumber', 'CustomerId', 'Surname'], inplace=True, axis=1)
```

```python
df
```

```python
df= pd.get_dummies(df,columns=['Geography', 'Gender'],drop_first=True)
df
```

```python
X=df.drop(['Exited'], axis=1)
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=4095)
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
model=Sequential()
model.add(Dense(30, activation='relu', input_dim=11))
model.add(Dense(25, activation='relu', input_dim=11))
model.add(Dense(15, activation='relu', input_dim=11))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
model.fit(X_train_scaled, y_train, epochs=40, validation_split=0.2)
```

```python
model.layers[0].get_weights()
```

```python
y_prob= model.predict(X_test_scaled)
y_prob
```

```python
y_pred= y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
